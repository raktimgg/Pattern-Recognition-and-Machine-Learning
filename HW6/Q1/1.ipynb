```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd           # Used only to read data
```

```python
data1 = pd.read_csv('datasets/Train_Tickets.csv')
```

```python
data2 = data1["Count"]
data = np.array(data2)
```

```python
# Plotting the histogram
plt.hist(data,bins = 100,label ='Histogram of given values',density = True)
plt.legend()
plt.show()
```

```python
# Clearly, it can be seen that the distribution is an exponential
# Since, the only parameter in exponential distribution is lambda, we can find it as
# lambda = N/sum(Xi)
```

```python
# Finding parameters of the distribution
lam = data.shape[0]/(np.sum(data))
print("Lambda = ",lam)
```

```python
# Plotting the fitted curve and comparing
ax = np.linspace(0,1200,1e4)
ay = lam*np.exp(-lam*ax)
plt.plot(ax,ay, label = 'PDF from ML estimate')
plt.hist(data,bins = 100,label ='Histogram of given values',density = True)
plt.legend()
plt.show()
```
