```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating data for sinusoid
x = np.random.rand(100)
e = np.random.normal(0,0.1,100)
x = np.sort(x)
y = np.sin(2*np.pi*x) + e
plt.plot(x,y)
plt.show()
```

```python
# Fitting using gaussian kernel
# Training
nos = 20 # Dimension of the kernel
mu = np.linspace(0,1,nos) # Array of means to be used in guassian kernel
sigma = 1 # Value of sigma in gaussian kernel
phi = np.empty([x.shape[0],nos])
for i in range(0,phi.shape[0]):
    for j in range(0,phi.shape[1]):
        phi[i,j] = np.exp(-((x[i]-mu[j])**2)/(2*sigma**2))
w = np.matmul(np.linalg.inv(np.matmul(phi.T,phi)),np.matmul(phi.T,y))
# Testing
y2 = np.zeros(y.shape)
for i in range(0,x.shape[0]):
    for j in range(0,w.shape[0]):
        y2[i]+=w[j]*np.exp(-((x[i]-mu[j])**2)/(2*sigma**2))
plt.plot(x,y2,label = 'Predicted curve', color = 'red')
plt.scatter(x, y, label = 'Test data labels')
plt.legend()
plt.show()
```
