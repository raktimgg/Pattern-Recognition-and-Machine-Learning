```python
#1.Solve using linear regression using a straight line.Plot the line and indicate the values of slope and intercept.
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = 10 * np.random.rand(50)
y = 2 * x - 5 + np.random.randn(50)
# plt.scatter(x, y)
# plt.show()
```

```python
#Solution to Question 1
phi = np.empty([x.shape[0],2])
for i in range(0,phi.shape[0]):
    phi[i,0] = 1.0
    phi[i,1] = x[i]
# print(phi.shape)
w = np.matmul(np.linalg.inv(np.matmul(phi.T,phi)),np.matmul(phi.T,y))
# print(w.shape)
x1 = np.linspace(0,10,1000)
y1 = w[0] + w[1]*x1
print('Slope = ',w[1])
print('Intercept = ',w[0])
plt.plot(x1,y1,label = 'Predicted line', color = 'red')
plt.scatter(x, y, label = 'Scatter points')
plt.legend()
plt.show()
```

```python
#2.Solve using linear regression using a polynomial basis.PLot the predicted curve.
import matplotlib.pyplot as plt
import numpy as np
x = 10 * np.random.rand(50)
y = np.sin(x) + 0.1 * np.random.rand(50)
# plt.scatter(x, y)
# plt.show() 
```

```python
# Solution to Question 2
degree = 11 # The maximum degree of the polynomial basis used
deg = degree-1 
phi = np.empty([x.shape[0],deg])
for i in range(0,phi.shape[0]):
    for j in range(0,phi.shape[1]):
        phi[i,j] = x[i]**j*1.0
# print(phi.shape)
w = np.matmul(np.linalg.inv(np.matmul(phi.T,phi)),np.matmul(phi.T,y))
print(w.shape)
x1 = np.linspace(0,10,1000)
y1 = np.zeros(x1.shape)
for i in range(0,x1.shape[0]):
    for j in range(0,w.shape[0]):
        y1[i]+=w[j]*x1[i]**j
plt.plot(x1,y1,label = 'Predicted curve', color = 'red')
plt.scatter(x, y, label = 'Scatter points')
plt.legend()
plt.show()
```

```python
# Solution to Question 3
nos = 10 # Dimension of the kernel
mu = np.linspace(0,nos,nos) # Array of means to be used in guassian kernel
sigma = 1 # Value of sigma in gaussian kernel
phi = np.empty([x.shape[0],nos])
for i in range(0,phi.shape[0]):
    for j in range(0,phi.shape[1]):
        phi[i,j] = np.exp(-((x[i]-mu[j])**2)/(2*sigma**2))
# print(phi.shape)
w = np.matmul(np.linalg.inv(np.matmul(phi.T,phi)),np.matmul(phi.T,y))
# print(w.shape)
x1 = np.linspace(0,10,1000)
y1 = np.zeros(x1.shape)
for i in range(0,x1.shape[0]):
    for j in range(0,w.shape[0]):
        y1[i]+=w[j]*np.exp(-((x1[i]-mu[j])**2)/(2*sigma**2))
plt.plot(x1,y1,label = 'Predicted curve', color = 'red')
plt.scatter(x, y, label = 'Scatter points')
plt.legend()
plt.show()
```

```python
# Solution to Question 6
# (a)
import numpy as np 
import matplotlib.pyplot as plt 

def getparameter(a,n):
	N = a.shape[0]
	return np.sum(a)*1.0/(n*N)

n = 1000
p = 0.5
size = 10000
x = np.random.binomial(n,p,size)
# print x.shape
param = getparameter(x,n)
print("From our definition of binomial, p = ", p)
print("From MLE, estimated value of p = ", param)

xn = np.random.binomial(n,param,size)

plt.hist(x,1000,label ='Original Sample',density = True,histtype = 'step')
plt.hist(xn,1000,label ='Sample from predicted parameter',density = True,histtype = 'step')
plt.legend()
plt.show()
```

```python
# (b)
import numpy as np
import matplotlib.pyplot as plt 

def getparam(a):
	N = a.shape[0]
	return np.sum(a)*1.0/N

lam = 1.0
size = 100
x = np.random.poisson(lam,size)
param = getparam(x)
print("Value of lambda given = ", lam)
print("Value of lambda by using MLE = ", param)
xn = np.random.poisson(param,size)

plt.hist(x,100,label ='Original Sample',density = True,histtype = 'step')
plt.hist(xn,100,label ='Sample from predicted parameter',density = True,histtype = 'step')
plt.legend()
plt.show()
```

```python
# (c)
import numpy as np
import matplotlib.pyplot as plt 

def getparam(a):
	N = a.shape[0]
	return (N*1.0)/np.sum(a)

lam = 2.0
scale = 1/lam
size = 10000
x = np.random.exponential(scale,size)    ################# The functions acceps scale as a parameter and not lambda ####################
param = getparam(x)
print("Value of lambda given = ", lam)
print("Value of lambda by using MLE = ", param)
s1 = 1/param
xn = np.random.exponential(s1,size)


plt.hist(x,1000,label ='Original Sample',density = True,histtype = 'step')
plt.hist(xn,1000,label ='Sample from predicted parameter',density = True,histtype = 'step')
plt.legend()
plt.show()
```

```python
# (d)
import numpy as np
import matplotlib.pyplot as plt 

def getparam(a):
	N = a.shape[0]
	mu1 = np.sum(a)*1.0/N
	temp = (a - mu1*np.full((a.shape[0]),1.))*(a - mu1*np.full((a.shape[0]),1.))
	sig1 = np.sum(temp)*1.0/N
	return mu1, sig1

mu = 0
sig = 2
size = 10000
x = np.random.normal(mu,sig,size)
mu_pred, sig_pred = getparam(x)
print("Given Value of mean = ", mu)
print("Using MLE, value of mean = ", mu_pred)
print("Given Value of standard deviation = ", sig*sig)
print("Using MLE, value of standard deviation = ", sig_pred)

xn = np.random.normal(mu,np.sqrt(sig_pred),size)


plt.hist(x,1000,label ='Original Sample',density = True,histtype = 'step')
plt.hist(xn,1000,label ='Sample from predicted parameter',density = True,histtype = 'step')
plt.legend()
plt.show()
```
