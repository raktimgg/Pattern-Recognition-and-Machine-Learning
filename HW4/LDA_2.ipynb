```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Reading data
data = np.loadtxt('datasets/WINE/wine.data',delimiter = ',')
np.random.shuffle(data)
```

```python
x = data[:,1:]
y_t = data[:,0]
```

```python
y = np.zeros([y_t.shape[0],3])
for i in range(0,y.shape[0]):
    if(y_t[i]==1):
        y[i][0] = 1
    elif(y_t[i]==2):
        y[i][1] = 1
    else:
        y[i][2] = 1
```

```python
## Creating features and labels in train and test set
x_train1 = x[:int(x.shape[0]*0.9)]
y_train1 = y[:int(x.shape[0]*0.9)]
x_test1 = x[int(x.shape[0]*0.9):]
y_test1 = y[int(x.shape[0]*0.9):]
print(x_train1.shape,y_train1.shape,x_test1.shape,y_test1.shape)
```

```python
x_train = x_train1
x_test = x_test1

# Uncomment the following lines to normalise data. However, the model is seen to perform better without normalisation

# Normalisation
# x_train = np.empty(x_train1.shape)
# m = []
# v = []
# for i in range(0,x_train1.shape[1]):
#     mean = np.mean(x_train1[:,i])
#     var = np.var(x_train1[:,i])
#     m.append(mean)
#     v.append(var)
#     x_train[:,i] = x_train1[:,i] - mean*np.ones(x_train1[:,i].shape)
#     x_train[:,i] = x_train[:,i]/var
y_train = y_train1
y_test = y_test1
```

```python
# Training
epochs = 30000  ## Number of epochs
w = np.zeros([x_train.shape[1],3])
lr = 0.000000001   ## Learning rate
batch_size = 160   ## Batch size
batch = int(x_train.shape[0]/batch_size)
# print(batch)
loss = []
for i in range(0,epochs):
    for b in range(0,batch):
        w= w - lr*(-2*(x_train[b:b+batch_size].T@(y_train[b:b+batch_size]-x_train[b:b+batch_size]@w)))
    y_pred = x_train@w
    l1 = (np.sum((y_train-x_train@w))**2)/y_train.shape[0]
    print("Epoch",i,"Loss",l1)
    loss.append(l1)
```

```python
## Plotting the loss
ax = np.arange(len(loss))
plt.plot(ax,loss)
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.show()
```

```python
## Predicting on test set and finding accuracy
print("Predicting on test set")
y_hat_t = (x_test@w)
loss = (np.sum((y_test-y_hat_t))**2)/y_test.shape[0]
print("Testing loss =",loss)
y_hat = np.zeros(y_hat_t.shape)
for i in range(0,y_hat.shape[0]):
    ind = np.argmax(y_hat_t[i])
    y_hat[i][ind] = 1
        
tot = 0.0
for i in range(0,y_hat.shape[0]):
    t = 0
    for j in range(0,y_hat.shape[1]):
        if(y_hat[i][j]!=y_test[i][j]):
            t = 1
    if(t==0):
        tot+=1
print("Testing accuracy =",(tot/y_hat.shape[0])*100,"%")
```
