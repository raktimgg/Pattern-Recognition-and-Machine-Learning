```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd ## Used only to read data
```

```python
data1 = pd.read_csv('datasets/IRIS/iris.data',header =None)
```

```python
data = data1
```

```python
data_l = data[4]
data_f = np.empty([data.shape[0],data.shape[1]-1])
for i in range(0,data_f.shape[1]):
    data_f[:,i] = data[i]
```

```python
y = np.zeros([data_l.shape[0],3])
for i in range(0,data_l.shape[0]):
    if(data_l[i]=='Iris-versicolor'):
        y[i][0] = 1
    elif(data_l[i]=='Iris-setosa'):
        y[i][1] = 1
    else:
        y[i][2] = 1
x = data_f
```

```python
## Creating train and test sets
x_train1 = np.empty([int(x.shape[0]*0.9),x.shape[1]])
y_train1 = np.empty([int(y.shape[0]*0.9),y.shape[1]])
x_test1 = np.empty([int(x.shape[0]*0.1),x.shape[1]])
y_test1 = np.empty([int(y.shape[0]*0.1),y.shape[1]])
print(x_train1.shape,y_train1.shape,x_test1.shape,y_test1.shape)
```

```python
## Creating features and labels in trian and test set
## Test set is made in such a way that it contains data corresponding to all the labels
x_train1[:45] = x[:45]
x_train1[45:90] = x[50:95]
x_train1[90:135] = x[100:145]

x_test1[:5] = x[45:50]
x_test1[5:10] = x[95:100]
x_test1[10:15] = x[145:150]

y_train1[:45] = y[:45]
y_train1[45:90] = y[50:95]
y_train1[90:135] = y[100:145]

y_test1[:5] = y[45:50]
y_test1[5:10] = y[95:100]
y_test1[10:15] = y[145:150]
```

```python
x_train = x_train1
y_train = y_train1
x_test = x_test1
y_test = y_test1

# Uncomment the following lines to normalise the data. However, the model is performing better without normalisation

# Normalisation
# x_train = np.empty(x_train1.shape)
# m = []
# v = []
# for i in range(0,x_train1.shape[1]):
#     mean = np.mean(x_train1[:,i])
#     var = np.var(x_train1[:,i])
#     m.append(mean)
#     v.append(var)
#     x_train[:,i] = x_train1[:,i] - mean*np.ones(x_train1[:,i].shape)
#     x_train[:,i] = x_train[:,i]/var
# y_train = y_train1
# y_test = y_test1
```

```python
# Training
epochs = 45000    ## Number of epochs
w = np.zeros([x_train.shape[1],3])
lr = 0.000001   ## Learning rate
batch_size = 135   ## Batch size
batch = int(x_train.shape[0]/batch_size)
# print(batch)
loss = []
for i in range(0,epochs):
    for b in range(0,batch):
        w= w - lr*(-2*(x_train[b:b+batch_size].T@(y_train[b:b+batch_size]-x_train[b:b+batch_size]@w)))
    y_pred = x_train@w
    l1 = (np.sum((y_train-x_train@w))**2)/y_train.shape[0]
    print("Epoch",i,"Loss",l1)
    loss.append(l1)
```

```python
## plotting the loss
ax = np.arange(len(loss))
plt.plot(ax,loss)
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.show()
```

```python
## Predicting on test set and finding accuracy
print("Predicting on test set")
y_hat_t = (x_test@w)
loss = (np.sum((y_test-y_hat_t))**2)/y_test.shape[0]
print("Testing loss =",loss)
y_hat = np.zeros(y_hat_t.shape)
for i in range(0,y_hat.shape[0]):
    ind = np.argmax(y_hat_t[i])
    y_hat[i][ind] = 1
        
tot = 0.0
for i in range(0,y_hat.shape[0]):
    t = 0
    for j in range(0,y_hat.shape[1]):
        if(y_hat[i][j]!=y_test[i][j]):
            t = 1
    if(t==0):
        tot+=1
print("Testing accuracy =",(tot/y_hat.shape[0])*100,"%")
```
